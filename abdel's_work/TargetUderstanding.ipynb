```python

import seaborn as sns


# Convert categorical feature to numerical values using one-hot encoding
df_encoded = pd.get_dummies(df, columns=['Failure Type'], prefix='Cat')

# Concatenate the original binary feature with the one-hot encoded categorical features
df_combined = pd.concat([df['Target'], df_encoded], axis=1)

# Calculate the correlation matrix
correlation_matrix = df_combined.corr()

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Set the title
plt.title('Heatmap of Correlation Between FeatureType1 and FeatureType2')

# Display the plot
plt.show()
```

```python

# Create a pivot table
pivot_table = df.pivot_table(index='Target', columns='Failure Type')

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', linewidths=0.5)

# Set the title
plt.title('Heatmap of Values Between FeatureType1 and FeatureType2')

# Display the plot
plt.show()
```

```python

# Get a list of all features (columns)
all_features = df.columns.tolist()

# Display the list of features
print("List of Features:", all_features)


```

```python

# Remove one or more features using drop()
features_to_remove = ['UDI', 'Product ID', 'Type', 'Air temperature [K]', 'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']
df_tarAndFai = df.drop(features_to_remove, axis=1)

# Display the modified DataFrame
print(df_tarAndFai)
```

```python

# Create a grouped bar chart
plt.figure(figsize=(10, 6))
sns.countplot(x='Failure Type', hue='Target', data=df_tarAndFai, palette='coolwarm')

# Set labels and title
plt.xlabel('Categorical Feature')
plt.ylabel('Count')
plt.title('Grouped Bar Chart of BinaryFeature by CategoricalFeature')
plt.xticks(rotation=45) 
# Display the plot
plt.show()
```

```python

# Create a pivot table and normalize values to percentages
pivot_table = pd.pivot_table(df_tarAndFai, index='Target', columns='Failure Type', aggfunc='size', fill_value=0)
normalized_pivot = pivot_table.div(pivot_table.sum(axis=1), axis=0) * 100

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(normalized_pivot, annot=True, cmap='Blues', fmt='.2f', linewidths=0.5)

# Set title
plt.title('Percentage Heatmap of BinaryFeature vs CategoricalFeature')

# Display the plot
plt.show()
```
